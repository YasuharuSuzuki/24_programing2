## 01. 都道府県データをカンマ区切りから空白区切りへ変換する

```python
# drive.mount()を使用してGoogleDriveをMount
# from google.colab import drive
# drive.mount('/content/drive')

# os.chdir()を使用してカレントディレクトリを変更。作成したショーットカット名を使ってパスを指定します。
# import os
# os.chdir('/content/drive/MyDrive/＜作成したショートカット名＞')

# データファイル定義
PREFECTURES_READ_COMMA_PATH = "./data/prefectures_comma.txt"
PREFECTURES_WRITE_PATH = "./data/prefectures.txt"

# ファイルを読み込みながら空白区切りに変換
prefectures2 = []
with open(PREFECTURES_READ_COMMA_PATH, mode="r", encoding="utf_8") as f:
    prefectures_all = f.readlines()
    prefectures_all2 = []
    for prefecture1 in prefectures_all:
        prefectures_all2.append(" ".join(prefecture1.rstrip().split(',')))
```

```python
# ファイル出力
with open(PREFECTURES_WRITE_PATH, mode="w", encoding="utf_8") as f:
    f.write("\n".join(prefectures_all2))
```

## 02. ワクチンデータを日付ｘ県ごとに集計し、カンマ区切りから空白区切りへ変換する

```python
# データの定義
VACCINATION_READ_COMMA_PATH = "./data/vaccination_comma.txt"
VACCINATION_WRITE_PATH = "./data/vaccination.txt"

# ファイルを読み込みながら空白区切りに変換
vaccination = []
with open(VACCINATION_READ_COMMA_PATH, mode="r", encoding="utf_8") as f:
    vaccination_lines = f.readlines()
    for vaccination1 in vaccination_lines:
        vaccination.append(vaccination1.rstrip().split(','))
```

```python
import pandas as pd

df_count1 = pd.DataFrame(vaccination, columns=['day','prefecture','count'])
df_count1['count'] = df_count1['count'].astype(int)
df_count1
```

```python
df_count2 = df_count1.groupby(['day','prefecture']).sum().reset_index()
df_count2
```

```python
# 出力するデータを作成する
vaccination_rows = []
def apply_func(row1):
    vaccination_rows.append(" ".join([row1['day'], row1['prefecture'], str(row1['count'])]))
df_count2.apply(apply_func, axis=1)

# ファイル出力
with open(VACCINATION_WRITE_PATH, mode="w", encoding="utf_8") as f:
    f.write("\n".join(vaccination_rows))
```
